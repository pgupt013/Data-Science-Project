# Reading the Air Quality Data

```python
dfaq=pd.read_excel("C:\\Users\\poonam\\Desktop\\FIU\\2019 Fall\\Data Science\\Project\\Data files\\annual_aqi_total.xlsx")
```

```python
dfaq1=dfaq[['State','Year','Unhealthy for Sensitive Groups Days','Unhealthy Days']]
dfaq2=dfaq1.groupby(['State','Year']).mean()
dfaq2[:2]
```

```python
dfaq5=dfaq2.groupby(['State']).sum()
dfaq5[:2]
```

```python
dfaq5.plot.bar(figsize=(12,5))
```

```python
#dfaq5.describe()
dfaq6=dfaq5
dfaq6.describe()
```

```python
dfaq7=dfaq6.drop(['Canada','Country Of Mexico','Guam','Puerto Rico','Virgin Islands'])
#dfaq7.describe()
```

```python
dfaq7=dfaq7.rename(index={'District Of Columbia':'District of Columbia'})
dfaq7[:2]
```

```python
totalbystate1=pd.DataFrame(totalbystate1)
totalbystate1[:2]
```

```python
temp1=dfaq7.join(totalbystate1,how='outer')
temp1[:2]
```

```python
temp1.columns = ['Unhealthy_for_Sensitive_Groups_Days','Unhealthy_Days','Total_CO2_emission']


### Cell 2
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax2 = ax.twinx()
width = 0.2
temp1.Unhealthy_for_Sensitive_Groups_Days.plot(kind='bar', color='red',ax=ax, width=width, position=1)
temp1.Unhealthy_Days.plot(kind='bar',color='green',ax=ax,width=width,position=2)
temp1.Total_CO2_emission.plot(kind='bar',ax=ax2,width=width,position=0)

plt.show()
```

# CO2 emission vs Uhealthy days

```python
temp2=temp1.sort_values(by=['Total_CO2_emission'],ascending=True)
#temp2[:5]
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111)
ax2 = ax.twinx()
width = 0.2
temp2.Unhealthy_for_Sensitive_Groups_Days.plot(kind='bar', color='orange',ax=ax, width=width, position=1,title="Unhealthy days vs CO2 emission")
temp2.Unhealthy_Days.plot(kind='bar',color='mediumseagreen',ax=ax,width=width,position=2)
temp2.Total_CO2_emission.plot(kind='bar',ax=ax2,width=width,position=0)

ax.set_xlabel('State')
ax.set_ylabel('Unhealthy days')
ax2.set_ylabel('Total CO2 emission')
ax.legend(loc = 'upper left')
ax2.legend()
plt.show()
```

# Linear regression model
# Relationship between CO2 emission and unhealthy days for sensitive group

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

plt.rcParams['figure.figsize'] = (12.0, 5.0)
x=totalbystate1.values.reshape(-1,1)
#print(x)
k1=temp1['Unhealthy_for_Sensitive_Groups_Days']
k=k1.values.reshape(-1,1)
#print(k)

x_train, x_test, k_train, k_test = train_test_split(x, k, test_size=0.4, random_state=1)
reg =LinearRegression()

reg.fit(x_train, k_train)
k_pred=reg.predict(x_test)
plt.scatter(x,k)
plt.plot(x_test,k_pred,color='red',linewidth=2)
plt.title('CO2 Emission vs Unhealthy days for Sensitive group')
plt.xlabel('CO2 Emission by fossil fuels')
plt.ylabel('Unhealthy days for Sensitive Groups')
plt.show()
```

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(x, k) #define the columns to perform calculations on
print("Pearson Correlation Coefficient: ", pearson_coef)
```

```python
stats.pearsonr(x, k)
```

```python
r2_score=reg.score(x,k)
print(r2_score)
```
