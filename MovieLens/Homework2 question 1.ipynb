### In this note book, we explored MovieLens dataset. Each movie in the database has a genre (comedy, animations, etc.) associated with it. we tried to solve the following:
#### 1. Top 20 genres with the highest number of responses from users.
#### 2. Top 20 genres sorted by average ratings.
#### 3. Top 20 movies sorted by descending mean female ratings for a specific genre (say \Drama").

```python
import pandas as pd
```

### Loading data from files

```python


# Data is from https://grouplens.org/datasets/movielens/
# Use "MovieLens 1M Dataset"
#dir = "C:\\Users\\poonam\\Desktop\\FIU\\2019 Fall\\Data Science\\Test coding folder\\Movie data\\ml-1m\\"
dir = "C:\\Users\\poonam\\Desktop\\FIU\\2019 Fall\\Data Science\\Home work\\HW2\\Submitted\\"
unames = ['user_id', 'gender', 'age', 'occupation', 'zip'] 
users = pd.read_csv(dir + 'users.dat', sep='::', header=None, names=unames,engine='python')
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv(dir + 'ratings.dat', sep='::', header=None, names=rnames, engine='python')
mnames = ['movie_id', 'title', 'genres']
#movies = pd.read_table(dir + 'ml-1m\\movies.dat', sep='::', header=None, names=mnames, engine='python')
movies = pd.read_csv(dir + 'movies.dat', sep='::', header=None, names=mnames, engine='python')
```

### Merging the data (SQL inner join)

```python
data = pd.merge(pd.merge(ratings,users),movies)
data[:5]
```

```python
data.size
```

###  Show the top 20 genres with the highest number of responses from users

```python
mean_ratings = data.pivot_table('rating', 'genres', aggfunc='count')
topgeneres = mean_ratings.sort_values(by=['rating'],ascending=False)
topgeneres[:20]
```

###  Top 20 genres sorted by average ratings

```python
mean_ratings = data.pivot_table('rating', 'genres', aggfunc='mean')
topgeneres = mean_ratings.sort_values(by=['rating'],ascending=False)
topgeneres[:20]
```

### Top 20 movies sorted by descending mean female ratings for a specific genre (say “Drama”).

```python
movieswithgenrecomedy = data[data['genres']== 'Comedy']
moviesratedbyfemales = movieswithgenrecomedy[movieswithgenrecomedy['gender'] == 'F']
topcomedymoviesratedbyfemales =  moviesratedbyfemales.pivot_table('rating', 'title')
topcomedymoviesratedbyfemales.sort_values(by=['rating'],ascending=False)[:20]
```
